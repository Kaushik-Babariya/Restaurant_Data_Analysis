```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
df
```

```python
data = df.dropna()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
cuisines = data['Cuisines'].str.split(',').explode()
top = cuisines.value_counts()
print('The top 3 most commom cuisins:')
top.head(3)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top.head(3).index,y=top.head(3).values,palette='Set2')
plt.title('Top 3 cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()
```

```python
restaurants = len(data)
# restaurants
percentages = (cuisines.value_counts()/restaurants)*100
percentages
```

```python
plt.figure(figsize=(10,6))
plt.pie(percentages.head(3).values,labels=percentages.head(3).index,autopct='%1.1f%%')
plt.xlabel('cuisine')
plt.ylabel('percentage')
plt.title('Percentage of Restaurants that serve each of the top cuisines ')
plt.show()
```
