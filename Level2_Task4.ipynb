```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv('Dataset .csv')
#df
```

```python
data = df.dropna()
```

```python
data.isnull().sum()
```

```python
chain_threshold = 3
restaurant_counts = data['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts >= chain_threshold].index.tolist()
restaurant_chains
```

```python
chain_data = data[data['Restaurant Name'].isin(restaurant_chains)]
chain_analysis = chain_data.groupby('Restaurant Name').agg({'Aggregate rating':'mean','Votes':'sum'}).reset_index()
print(chain_analysis.sort_values(by='Aggregate rating',ascending=False))
```
