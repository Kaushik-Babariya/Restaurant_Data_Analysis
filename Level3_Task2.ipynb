```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv('Dataset .csv')
#df
```

```python
data = df.dropna()
```

```python
data.isnull().sum()
```

```python
highest_votes = data[data['Votes']==data['Votes'].max()]
print('Highest Votes')
print(highest_votes[['Restaurant Name','Votes']])
```

```python
lowest_votes = data[data['Votes']==data['Votes'].min()]
print('Lowest Votes')
print(lowest_votes[['Restaurant Name','Votes']])
```

```python
from scipy import stats
```

```python
corre = stats.pearsonr(data['Votes'],data['Aggregate rating'])
corre
```

```python
plt.figure(figsize=(10,5))
sns.regplot(x='Votes',y='Aggregate rating',data=data)

plt.show()
```
