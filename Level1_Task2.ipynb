```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
df
```

```python
data = df.dropna()
```

```python
data.isnull().sum()
```

```python
top10city = data['City'].value_counts().head(10)
top10city
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top10city.index,y=top10city.values,palette='Set2',legend='auto')
plt.xticks(rotation=50)
plt.xlabel('City')
plt.ylabel('Count of restaurants')
plt.title('Top 10 City with The Highest Number Of Restaurants')
plt.show()
```

```python
city = data.groupby('City')['Aggregate rating'].mean()
```

```python
highest_rating_cities = city.sort_values(ascending=False)
highest_rating_cities
```

```python
print("the city with highest average Rating :")
high_city = highest_rating_cities.idxmax()
high_city
```

```python
plt.figure(figsize=(10,5))
plt.bar(highest_rating_cities.head(10).index,highest_rating_cities.head(10).values)
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Top 10 City with The Highest Average Rating')
plt.xticks(rotation=45)
plt.show()
```
