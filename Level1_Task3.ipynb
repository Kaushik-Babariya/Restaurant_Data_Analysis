```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
#df
```

```python
data = df.dropna()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data['Price range'],data=data,palette='Set2')
plt.title('Distribution of Price Ranges among The Restaurants')
plt.show()
```

```python
percentage_per_pricerange = (data['Price range'].value_counts()/len(data))*100
print("The percentage of restaurantsin each price range category.")
percentage_per_pricerange
```

```python
plt.figure(figsize=(10,5))
plt.bar(percentage_per_pricerange.index,percentage_per_pricerange.values)
plt.xlabel('Price Range')
plt.ylabel('Percentages of each price range')
plt.title('')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(percentage_per_pricerange.values,labels=percentage_per_pricerange.index,autopct='%1.1f%%')
plt.title('Percentage of Restaurants in each Price Range Category')
plt.legend()
plt.show()
```
